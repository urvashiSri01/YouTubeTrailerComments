```python
import pandas as pd
import numpy as np
import regex as re
```

```python
with open("YouTubeTrailerComments.txt") as f:
    data = f.read()
```

```python
splitted_data = data.splitlines()
```

```python
splitted_data
```

```python
df = pd.DataFrame(columns=['id','vc','lc','dc','fc','cc'])
```

```python
movie_id = 0 
movie_reviews = 0

for data in splitted_data:
    temp1 = re.findall('MovieID::::[0-9]+', data)
    temp2 = re.findall('^[0-9]+\s+[0-9]+\s+[0-9]+\s[0-9]+\s[0-9]+', data)
    
    if temp1 != []:
        movie_id = str(temp1[0]).replace('MovieID::::', '')
        
    if temp2 != []:
        movie_reviews = str(temp2[0]).split(" ")
        
    
    if movie_id != 0 and movie_reviews != 0:
        df.loc[movie_id] = {'id': movie_id, 'vc':movie_reviews[0],'lc':movie_reviews[1],
                               'dc':movie_reviews[2],'fc':movie_reviews[3],'cc':movie_reviews[4]}
        
        movie_id = movie_reviews = 0
```

```python
df.to_csv("out.csv")
```
